# Fibonacci Series

```python
Number = int(input("\nPlease Enter the Range Number: "))
i = 0
FirstValue = 0
SecondValue = 1
while(i < Number):
    if(i <= 1):
        Next = i
    else:
        Next = FirstValue + SecondValue
        FirstValue = SecondValue
        SecondValue = Next
    print(Next)
    i = i + 1
```
