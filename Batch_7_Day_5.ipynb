# Prime Number 

```python
nums = range(2, 2500) 
for i in range(2, 2500): 
     nums = filter(lambda x: x == i or x % i, nums)
 
print(nums)

def isPrime(N): 
  
    i, j, flag = 0, 0, 0; 
  
    print("Prime Numbers are:"); 
  
    for i in range(1, N + 1, 1): 
  
        if (i == 1 or i == 0): 
            continue; 
  
        flag = 1; 
  
        for j in range(2, ((i // 2) + 1), 1): 
            if (i % j == 0): 
                flag = 0; 
                break; 
  
        if (flag == 1): 
            print(i, end = "\t"); 
  
N = 2500; 
isPrime(N); 
  
```

#  Lambda Function for Capitalizing the whole sentence

```python
a = str(input("Input:-"))
list(map(lambda x: x.capitalize(), [a]))
a.title()
```
