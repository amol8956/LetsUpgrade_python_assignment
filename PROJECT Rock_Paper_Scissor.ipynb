# SILVER PROJECT: Rock - Paper - Scissor

```python
import random
def player_choice(user_choice):
    if user_choice == 1:
        print("You chose:   rock\n")
    elif user_choice == 2:
        print("You chose:    paper\n")
    else:
        print("You chose:   scissors\n")

def computer_choice(cpu_choice):
    if cpu_choice == 1:
        print("The computer chose:   rock\n")
    elif cpu_choice == 2:
        print("The computer chose:   paper\n")
    else:
        print("The computer chose:   scissors\n")

def result(user_choice, cpu_choice, player_score, cpu_score):
    if user_choice == cpu_choice:
        print("Game Tie! This Match point has been divided by 50% to each side...\n")
        return [player_score + 0.5, cpu_score + 0.5]
    elif user_choice == 1 and cpu_choice == 3:
        print("You Win! 1 point added to your point table...\n")
        return [player_score + 1, cpu_score]
    elif user_choice == 2 and cpu_choice == 1:
        print("You Win! 1 point added to your point table...\n")
        return [player_score + 1, cpu_score]
    elif user_choice == 3 and cpu_choice == 2:
        print("You Win! 1 point added to your point table...\n")
        return [player_score + 1, cpu_score]
    else:
        print("Computer Win! 1 point added to Computer point table...\n")
        return [player_score, cpu_score + 1]


def print_score(p_score, c_score):
    print("Score:""\nYour Score: ", p_score,"\n" "\nComputer Score: ", c_score)

def validation_input():
    while True:
        try:
            user_input = int(input("Enter your choice:\t"))
            if user_input not in range(1, 4):
                print("Enter number between 1 to 3, Please try again\n")
                continue
            if type(user_input) == int:
                break
        except ValueError:
            print("Only accept exact numbers\n")
            continue
    return user_input


print("\n 1 - Rock\n 2 - Paper\n 3 - Scissors\n")
human_score = 0
computer_score = 0
while True:
    user = validation_input()
    player_choice(user)
    ai = random.randint(1, 3)
    computer_choice(ai)
    human_score, computer_score = result(user, ai, human_score, computer_score)
    print_score(human_score, computer_score)
    command = int(input("Type 0 to Exit the program, type any another number to continue this game\n"))
    if command == 0:
        break
```
